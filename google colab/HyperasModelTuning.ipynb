```
def model(x_train, y_train, x_test, y_test):  

    clear_session()

    model = Sequential()

    model.add(Conv2D({{choice([16, 32, 64])}}, kernel_size={{choice([(3, 3),(5,5)])}}, strides=1, activation='relu', input_shape=(100, 100, 3)))
    model.add(Conv2D({{choice([16, 32, 64])}}, kernel_size={{choice([(3, 3),(5,5)])}}, strides=2, activation='relu'))
    model.add(Dropout({{uniform(0, 1)}}))
    model.add(Conv2D({{choice([32, 64, 128])}}, kernel_size={{choice([(3, 3),(5,5)])}}, strides=1, activation='relu'))
    model.add(Conv2D({{choice([32, 64, 128])}}, kernel_size={{choice([(3, 3),(5,5)])}}, strides=2, activation='relu'))
    model.add(Dropout({{uniform(0, 1)}}))
    model.add(Conv2D({{choice([64, 128, 256])}}, kernel_size={{choice([(3, 3),(5,5)])}}, strides=1, activation='relu'))
    model.add(Conv2D({{choice([64, 128, 256])}}, kernel_size={{choice([(3, 3),(5,5)])}}, strides=2, activation='relu'))
    model.add(Flatten())
    model.add(Dropout({{uniform(0, 1)}}))
    model.add(Dense(512, activation='tanh'))
    model.add(Dense(24, activation='softmax'))

    model.summary()

    model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
    
    callback = EarlyStopping(monitor='val_loss', patience=3)

    result = model.fit(X_train, y_train, epochs={{choice([10,20,30])}}, verbose=2, validation_split=0.1, callbacks=[callback])
    
    #get the highest validation accuracy of the training epochs
    validation_acc = np.amax(result.history['val_accuracy']) 
    print('Best validation acc of epoch:', validation_acc)

    return {'loss': -validation_acc, 'status': STATUS_OK, 'model': model}
```

**Run the HYPERAS optimization process**

```
# Get best parameters to train model
best_run, best_model = optim.minimize(model=model,
                                      data=data,
                                      algo=tpe.suggest,
                                      max_evals=5, # should try to use 3 originally 5
                                      trials=Trials(),
                                      notebook_name='HyperasModelTuning')

X_train, Y_train, X_test, Y_test = data()
print("Evalutation of best performing model:")
print(best_model.evaluate(X_test, Y_test))
print("Best performing model chosen hyper-parameters:")
print(best_run)
```
